Data Cleaning & Pre-processing

```python
#dropping unnecessary column
df.drop(['_c39', 'policy_number', 'insured_zip'], axis=1, inplace=True)
df.head()
```

```python
#Convert Date Columns to datetime
df['policy_bind_date'] = pd.to_datetime(df['policy_bind_date'])
df['incident_date'] = pd.to_datetime(df['incident_date'])
```

```python
# Correcting negative or odd values 
df['umbrella_limit'] = df['umbrella_limit'].apply(lambda x: np.nan if x < 0 else x)
df['capital-gains'] = df['capital-gains'].apply(lambda x: np.nan if x < 0 else x)
df['capital-loss'] = df['capital-loss'].apply(lambda x: np.nan if x < 0 else x)
```

```python
#checking for missing values
missing_data = df.isnull().sum()
print(missing_data[missing_data > 0]) 
```

feature engeneering

time base features:
The number of days between the policy_bind_date and the incident_date can indicate how long a customer had the policy before the incident.

```python
df['policy_age'] = (df['incident_date'] - df['policy_bind_date']).dt.days
df.head()
```

```python
df['policy_age'].unique
```

Claim Ratio:
calculate ratios between various claims like injury_claim, property_claim, and vehicle_claim to see if the relationship between these claims is suspicious.

```python
df['injury_claim_ratio'] = df['injury_claim'] / df['total_claim_amount']
df['property_claim_ratio'] = df['property_claim'] / df['total_claim_amount']
df['vehicle_claim_ratio'] = df['vehicle_claim'] / df['total_claim_amount']
```

Total Time of Incident:

Use incident_hour_of_the_day to create new time segments, such as "Morning," "Afternoon," "Evening," and "Night". Certain time segments may correlate with fraudulent behavior.

```python
def time_of_day(hour):
    if 5 <= hour < 12:
        return 'Morning'
    elif 12 <= hour < 17:
        return 'Afternoon'
    elif 17 <= hour < 21:
        return 'Evening'
    else:
        return 'Night'

df['incident_time_of_day'] = df['incident_hour_of_the_day'].apply(time_of_day)
```

calculates how long ago the incident happened relative to the last data point we have. This could help capture potential outliers in terms of delayed reporting.

```python
df['days_since_incident'] = (pd.to_datetime('today') - df['incident_date']).dt.days
```

Data Normalization

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df[['policy_annual_premium', 'total_claim_amount', 'injury_claim']] = scaler.fit_transform(df[['policy_annual_premium', 'total_claim_amount', 'injury_claim']])
```
